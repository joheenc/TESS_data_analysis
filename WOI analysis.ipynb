```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("woidata_all.csv")
df.head
```

```python
plt.xlabel("log(P)")
plt.ylabel("log(Î¶)")
plt.scatter(np.log(df["Guessed_Period"]), np.log(df["Peak_Zeta"]), alpha=0.5, s=8, color="gray")
plt.show()
```

```python
plt.xlabel("log(P)")
plt.ylabel("log(chi2)")
plt.scatter(np.log(df["Guessed_Period"]), np.log(df["Chi2"]), alpha=0.5, s=8, color="gray")
plt.show()
```

```python
plt.xlabel("log(P)")
plt.ylabel("log(delt_chi2)")
plt.scatter(np.log(df["Guessed_Period"]), np.log(df["Delt_chi2"]), alpha=0.5, s=8, color="gray")
plt.show()
```

```python
plt.xlabel("log(P)")
plt.ylabel("log(kurtosis)")
plt.scatter(np.log(df["Guessed_Period"]), np.log(df["Kurtosis"]), alpha=0.5, s=8, color="gray")
plt.show()
```

```python
plt.xlabel("log(delt_chi2)")
plt.ylabel("log(kurtosis)")
plt.scatter(np.log(df["Delt_chi2"]), np.log(df["Kurtosis"]), alpha=0.5, s=8, color="gray")
plt.show()
```

```python
plt.xlabel("log(Î¶)")
plt.ylabel("log(Î¶)")
plt.scatter(np.log(df["Peak_Zeta"]), np.log(df["Delt_chi2"]), alpha=0.5, s=8, color="gray")
plt.show()
```

```python
plt.xlabel("log(Î¶)")
plt.ylabel("log(Kurtosis)")
plt.scatter(np.log(df["Peak_Zeta"]), np.log(df["Kurtosis"]), alpha=0.5, s=8, color="gray")
plt.show()
```
