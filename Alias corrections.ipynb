```julia
# include("sebastien.jl")
# using Main.WeirdDetector
# using PyPlot
using Statistics

lines = readlines("harmonics_corrected_round2.csv")
tid = parse(Int64, split(lines[2], ",")[1])
alias = parse(Float64, split(lines[2], ",")[2])
period = parse(Float64, split(lines[2], ",")[4])
df = loadFITS(0, fitsdir="fitsfiles/", tic_id=tid, detrend_kw=1)
scatter(df[:t] .% (period/alias), df[:F], alpha=0.4, s=0.1)
plot([0, period/alias], [Statistics.mean(df[:F]), Statistics.mean(df[:F])])
show()
```

```julia
#auto-get avged flux & duration
avgedfluxes = []
avgedts = []
nbins = 500
tstep = period/alias/nbins
duration = 0
df[:modt] = df[:t] .% (period/alias)
for currstart in 0:period/alias/nbins:period/alias
    currend = currstart + period/alias/nbins
    indices = findall(t -> t>=currstart && t<currend, df[:modt])
    if length(indices) > 0
        append!(avgedfluxes, Statistics.median(df[:F][indices]))
        append!(avgedts, Statistics.median(df[:modt][indices]))
    end
end
scatter(df[:t] .% (period/alias), df[:F], alpha=0.4, s=0.1)
scatter(avgedts, avgedfluxes, alpha=1, s=10, c="black")
show()

#duration calculation
idx = Statistics.argmin(avgedfluxes)
while avgedfluxes[idx] < Statistics.median(df[:F])
    duration += tstep
    idx -= 1
    if idx < 0
        idx = length(avgedts)
    end
end
idx = Statistics.argmin(avgedfluxes)
while avgedfluxes[idx] < Statistics.median(df[:F])
    duration += tstep
    idx += 1
    if idx > length(avgedts)
        idx = 1
    end
end
print(duration)
```

```julia
# using PyPlot
# using DataFrames
# using CSV
using Statistics
# include("sebastien.jl")
# using Main.WeirdDetector
data = CSV.read("harmonics_corrected_round2.csv")
durations = []
taus = []
depths = []
c = 1
for i in 1:size(data, 1)
    tid = data[i, :][1]
    alias = data[i, :][2]
    period = data[i, :][4]
    
    df = loadFITS(0, fitsdir="fitsfiles/", tic_id=tid, detrend_kw=1)
    #auto-get avged flux & duration
    avgedfluxes = []
    avgedts = []
    nbins = 500
    tstep = period/alias/nbins
    duration = 0
    df[:modt] = df[:t] .% (period/alias)
    for currstart in 0:period/alias/nbins:period/alias
        currend = currstart + period/alias/nbins
        indices = findall(t -> t>=currstart && t<currend, df[:modt])
        if length(indices) > 0
            append!(avgedfluxes, Statistics.median(df[:F][indices]))
            append!(avgedts, Statistics.median(df[:modt][indices]))
        end
    end
    scatter(df[:t] .% (period/alias), df[:F], alpha=0.4, s=0.1)
    scatter(avgedts, avgedfluxes, alpha=1, s=10, c="black")
    show()
    clf()

    #duration calculation
    idx = Statistics.argmin(avgedfluxes)
    while avgedfluxes[idx] < Statistics.median(df[:F])
        duration += tstep
        idx -= 1
        if idx < 1
            idx = length(avgedts)
        end
    end
    idx = Statistics.argmin(avgedfluxes)
    while avgedfluxes[idx] < Statistics.median(df[:F])
        duration += tstep
        idx += 1
        if idx > length(avgedts)
            idx = 1
        end
    end
    acc = 0
    onepd = 1
    while acc < period
        acc += df[:t][onepd+1]-df[:t][onepd]
        onepd += 1
    end
    tau0 = df[:t][Statistics.argmin(df[:F][1:onepd])] - duration*0.001388889
    depth = Statistics.median(avgedfluxes) - Statistics.minimum(avgedfluxes)
    append!(durations, duration)
    append!(taus, tau0)
    append!(depths, depth)
    println(depth)
    println(c)
    c += 1
end
```
